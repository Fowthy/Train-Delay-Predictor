```python
from IPython.display import Image #used for displaying images
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
import datetime
from dateutil import parser #used for parsing datetime columns
sns.set_theme()    #used for changing image sizes of plots, graphs etc.
```

```python
dfJanuarySchiphol = pd.read_csv('January_Schiphol_Cleaned.csv')
```

```python
dfJanuarySchiphol
```

```python
dfJanuarySchiphol = dfJanuarySchiphol.drop(columns = ['DepartureDatePlanned', 'DepartureTimePlanned', 'DepartureDateActual', 'DepartureTimeActual'])
```

```python
dfJanuarySchiphol
```

```python
dfJanuarySchiphol.dtypes
```

```python
dfJanuarySchiphol = dfJanuarySchiphol.drop(columns='Unnamed: 0')
```

```python
dfJanuarySchiphol['Train_Departure_Planned_Time'] = dfJanuarySchiphol.apply(lambda row: parser.parse(row.Train_Departure_Planned).time(), axis=1)
```

```python
dfJanuarySchiphol
```

```python
dfJanuarySchiphol["Train_Departure_Planned_Time"] = dfJanuarySchiphol["Train_Departure_Planned_Time"].astype(str)
```

```python
dfJanuarySchiphol['Train_Departure_Planned_Time'] = dfJanuarySchiphol['Train_Departure_Planned_Time'].str.replace(':', '', regex=False) 
```

```python
dfJanuarySchiphol.dtypes
```

```python
dfJanuarySchiphol["Train_Departure_Planned_Time"] = dfJanuarySchiphol["Train_Departure_Planned_Time"].astype(int)
```

```python
dfJanuarySchiphol
```

```python
# create Time groups
conditions = [
    #Night
    (dfJanuarySchiphol['Train_Departure_Planned_Time'] > 190000) & (dfJanuarySchiphol['Train_Departure_Planned_Time'] <= 240000),
    
    #Early Morning
    (dfJanuarySchiphol['Train_Departure_Planned_Time'] >= 0) & (dfJanuarySchiphol['Train_Departure_Planned_Time'] < 60000),
    
    #Morning Rush Hour
    (dfJanuarySchiphol['Train_Departure_Planned_Time'] >= 60000) & (dfJanuarySchiphol['Train_Departure_Planned_Time'] <= 100000),
    
    #Between Rush hours
    (dfJanuarySchiphol['Train_Departure_Planned_Time'] > 100000) & (dfJanuarySchiphol['Train_Departure_Planned_Time'] < 150000),
    
    #Afternoon Rush hour
    (dfJanuarySchiphol['Train_Departure_Planned_Time'] >= 150000) & (dfJanuarySchiphol['Train_Departure_Planned_Time'] <= 190000)
]

# create a list of the values we want to assign for each condition
values = ['Night', 'Early_Morning', 'Morning_Rush_Hour', 'Between_Rush_hours', 'Afternoon_Rush_hour']

# create a new column and use np.select to assign values to it using our lists as arguments
dfJanuarySchiphol['Time_Group'] = np.select(conditions, values)

# display updated DataFrame
dfJanuarySchiphol.head()
```

```python
dfJanuarySchiphol.to_csv('January_Schiphol_Time_Groups.csv')
```
