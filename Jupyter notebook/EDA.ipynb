```python
#here i load the libraries
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
import datetime
sns.set_theme()
```

```python
January=pd.read_csv('January.csv')
January.drop('Unnamed: 0',axis=1,inplace=True)
January.info()
January
```

```python
#First I make a date from the date planned departure, then i make it a string in a new column fo i can split it into 2 columns 
#and make the date alone a date
January['Train Planned Departure']=pd.to_datetime(January['Train Planned Departure'], format="%Y-%m-%dT%H:%M:%SZ")
January['Train Planned Departure1']=January['Train Planned Departure'].astype(str)
January[['DepartureDate','DepartureTime']]= January['Train Planned Departure1'].str.split(' ', expand = True)
January['DepartureDate']=pd.to_datetime(January['DepartureDate'])
January
```

```python
#then i make the depature time a time delta (Actually want to make it a normal time format)
January['DepartureTime']=pd.to_timedelta(January['DepartureTime'].str.strip())
print(January.info())
January
```

```python
#Then i also put the delay in time delta format
January['Train Delay']=pd.to_timedelta(January['Train Delay'].str.strip())
January
```
